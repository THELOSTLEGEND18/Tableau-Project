```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver = webdriver.Firefox()
```

```python
driver.get("https://www.fitternity.com/mumbai/malad-west/fitness")
driver.implicitly_wait(5)

button = driver.find_elements(By.CLASS_NAME, "fit-budget")
button[1].click()

driver.implicitly_wait(5)

button = driver.find_elements(By.CLASS_NAME, "fit-budget")
button[2].click()

driver.implicitly_wait(5)
```

```python
names = driver.find_elements(By.CLASS_NAME, "vendorname-span")
ratings = driver.find_elements(By.CLASS_NAME, "vendor-rating")
locations = driver.find_elements(By.CLASS_NAME, "vendor-location")
reviews = driver.find_elements(By.CLASS_NAME, "vendor-reviews")

df = pd.DataFrame(columns=["Name", "Rating", "Location", "Reviews"])

for i in range(len(names)):

    # If there are no reviews, set review to 0
    if reviews[i].text == "No reviews":
        review = 0
    else:
        review = reviews[i].text.split(" ")[0]

    # If there are no ratings, set rating to None
    if ratings[i].text == "- -":
        rating = None
    else:
        rating = ratings[i].text

    # Remove + n Other Location
    location = locations[i].text.strip("\ue55f")
    if location.endswith("Other Location"):
        location = location.split("+")[0]

    df = df._append({"Name": names[i].text, "Rating": rating, "Location": location, "Reviews": review}, ignore_index=True)

df["Reviews"] = df["Reviews"].astype(int)
df["Rating"] = df["Rating"].astype(float)
```
